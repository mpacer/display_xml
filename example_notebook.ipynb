# Example Notebook for display_xml

By default the output is folded. 

```python
from display_xml import XML
XML('<body><tag>content</tag></body>')
```

To expand all elements by default, use `expand=True`.

```python
XML('<body><tag>content</tag></body>', expand=True)  
```

Set `pretty_print` to control the width of the indentation level. The default is 1


```python
XML('<body><tag>content</tag></body>', expand=True, pretty_print=0)  
```

```python
XML('<body><tag>content</tag></body>', expand=True, pretty_print=3)  
```

## Styles

```python
XML.display_all_styles('<body><tag>content</tag></body>', expand=True)
```

```python
XML('<body><tag>content</tag></body>', expand=True, style="perldoc")
```
